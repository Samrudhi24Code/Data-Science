```python
# Importing essential libraries for data manipulation, visualization, scaling, dimensionality reduction, clustering, and evaluation
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical operations and working with arrays
import matplotlib.pyplot as plt  # For data visualization and plotting
import seaborn as sns  # For advanced data visualization and aesthetic plots
```

```python
data = pd.read_csv('homeprices.csv')  
print(data.head())  # Display the first few rows of the dataset
```

```python

# Check for null values

print(data.isnull().sum())#If the count for any column is greater than 0, that column contains missing values
#t Hence,here are no missing values in the dataset
```

# Standardization

```python
# Print the columns to check their names
print(data.columns)

# Replace 'price' with the actual name if it's different
X = data.drop(columns=['price'])  # Features (independent variables)
y = data['price']  # Target (dependent variable)

# Scale the features (as shown previously)
from sklearn.preprocessing import StandardScaler

# Convert categorical column to numerical if needed
X = pd.get_dummies(X, columns=['town'], drop_first=True)

# Scale only the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# Convert scaled data back to DataFrame if needed
scaled_data_df = pd.DataFrame(scaled_data, columns=X.columns)
```

# 2.2 Data Exploratory

```python
# Univariate Analysis means plotting histogram for each features.
data.hist(bins=20, figsize=(20, 15))
plt.show()
```

```python
# Bivariate Analysis
sns.pairplot(data)
plt.show()
```

# Step 3: Hierarchical and K-Means Clustering (Before PCA)



```python
# 3.1 Hierarchical Clustering on original data
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(scaled_data, 'ward')
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram (Before PCA)')
plt.show()
```

```python
# 3.2 K-Means Clustering on original data
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

```

```python
# Elbow plot for K-Means (Before PCA)
plt.plot(range(1, 10), inertia, marker='o')
plt.title('Elbow Method for Optimal K (Before PCA)')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# K-Means Clustering with k=3 
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster_KMeans_Original'] = kmeans.fit_predict(scaled_data)
data['Cluster_KMeans_Original']
```

# Step 4: PCA Analysis

```python
#PCA (Principal Component Analysis) is a dimensionality reduction technique used to simplify complex
#datasets by transforming them into fewer, uncorrelated components (called principal components) 
#while retaining most of the original data's variance.
#The idea is to reduce the number of features or variables, making the data easier to
#visualize and work with, without losing important information.
```

```python
# 4.1 Apply PCA to reduce the dataset to 3 components
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca_data = pca.fit_transform(scaled_data)
data['Cluster_KMeans_Original1'] = kmeans.fit_predict(pca_data)
```

# Step 5: Hierarchical and K-Means Clustering (After PCA)


```python
linked_pca = linkage(pca_data, 'ward')
plt.figure(figsize=(10, 7))
dendrogram(linked_pca)
plt.title('Hierarchical Clustering Dendrogram (After PCA)')
plt.show()
```

```python
# 5.2 K-Means Clustering on PCA-transformed data
inertia_pca = []
for k in range(1, 10):
    kmeans_pca = KMeans(n_clusters=k, random_state=42)
    kmeans_pca.fit(pca_data)
    inertia_pca.append(kmeans_pca.inertia_)

```

```python
# Elbow plot for K-Means on PCA-transformed data
plt.plot(range(1, 10), inertia_pca, marker='o')
plt.title('Elbow Method for Optimal K (After PCA)')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# K-Means Clustering on PCA data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
data['Cluster_KMeans_PCA'] = kmeans_pca.fit_predict(pca_data)
```

By implementing K-Means clustering on PCA-transformed data, you can derive actionable insights that enhance your understanding of the property market. This understanding can lead to better decision-making, more effective pricing strategies, and improved predictive models. The analysis not only aids in comprehending current market dynamics but also assists in forecasting future trends, making it a valuable tool in property data analysis.

```python
# Step 6: Compare Results
from sklearn.metrics import silhouette_score
# Silhouette Score for original K-Means clusters
silhouette_before = silhouette_score(scaled_data, data['Cluster_KMeans_Original'])
print(f'Silhouette Score (Before PCA): {silhouette_before}')

# Silhouette Score for PCA-based K-Means clusters
silhouette_after = silhouette_score(pca_data, data['Cluster_KMeans_PCA'])
print(f'Silhouette Score (After PCA): {silhouette_after}')
```

The observations indicate that the clustering quality is moderate with no significant improvement from PCA. Further analysis of clusters and exploration of other clustering methods or feature engineering might lead to better-defined clusters and insights from the dataset.
